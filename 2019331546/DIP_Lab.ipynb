```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Load images
img1 = cv2.imread("/content/sample_data/P1.png")
img2 = cv2.imread("/content/sample_data/P1.png")
```

```python
cv2_imshow(img1)
cv2_imshow(img2)
```

Resize Images

```python
img1=cv2.resize(img1,(300,300))
img2=cv2.resize(img2,(300,300))
```

Image addition

```python
img_add=cv2.add(img1,img2)
cv2_imshow(img_add)
```

Image Subtraction

```python
img_sub=cv2.subtract(img1,img2)
cv2_imshow(img_sub)
```

Image Multiplication

```python
img_mul = cv2.multiply(img1, img2)
cv2_imshow(img_mul)
```

Noise Reduction

```python
# 1. Noise Reduction
# Add Gaussian noise
noise = np.random.normal(0, 25, img1.shape).astype(np.uint8)
noisy_img = cv2.add(img1, noise)
cv2_imshow(noisy_img)

# Reduce noise using Gaussian Blur
blurred_img = cv2.GaussianBlur(noisy_img, (5, 5), 0)
cv2_imshow(blurred_img)
```

Change Detection

```python
# 2. Change Detection
# Simulate changes by adding noise to the second image
changed_img = cv2.add(img2, noise)
cv2_imshow(changed_img)

# Detect changes
change_mask = cv2.absdiff(img1, changed_img)
gray_diff = cv2.cvtColor(change_mask, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(gray_diff, 30, 255, cv2.THRESH_BINARY)
cv2_imshow(thresh)
```

Masking

```python
# 3. Masking
# Create a circular mask
mask = np.zeros(img1.shape[:2], dtype=np.uint8)
cv2.circle(mask, (mask.shape[1]//2, mask.shape[0]//2), 100, (255), -1)

# Apply mask
masked_img = cv2.bitwise_and(img1, img1, mask=mask)
cv2_imshow(mask)
cv2_imshow(masked_img)
```
